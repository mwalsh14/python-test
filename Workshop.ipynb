```python
def fahr_to_kelvin(temp):
    temp_k = ((temp - 32) * (5/9)) + 273.15
    return temp_k
```

```python
fahr_to_kelvin(32)
```

```python
def kelvin_to_celsius(temp_k):
    temp_c = temp_k - 273.15
    return temp_c
```

```python
kelvin_to_celsius(273.15)
```

```python
def fahr_to_celsius(temp):
    temp_c = kelvin_to_celsius(fahr_to_kelvin(temp))
    return temp_c
```

```python
fahr_to_celsius(32)
```
